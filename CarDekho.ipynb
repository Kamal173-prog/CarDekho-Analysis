```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams['figure.figsize']=(12,8)
```

```python
df=pd.read_csv("CAR_DETAILS_FROM_CAR_DEKHO.csv")
df.head()
```

```python
sns.countplot(x="fuel",hue='fuel',data=df,palette='Set2')
plt.title("Fuel distribution type")
plt.show()
```

```python
sns.histplot(df['selling_price'],kde=True,bins=30,color='skyblue')
plt.title("Selling price distribution")
plt.xlabel("Price (INR)")
plt.show()
```

```python
sns.barplot(x='fuel',hue='fuel',y='selling_price',data=df,palette='Set1')
plt.title("Price  By The Fuel Type")
plt.show()
```

```python
sns.boxplot(x='transmission',hue='transmission',y='selling_price',data=df,palette='coolwarm')
plt.title("Selling price by Transmission")
plt.show()
```

```python
sns.violinplot(x='owner',hue='owner',y='selling_price',data=df,palette="Set3")
plt.title("Selling price by owner")
plt.show()
```

```python
correlation=df[['year','km_driven','selling_price']].corr()
sns.heatmap(correlation,annot=True,cmap='YlGnBu')
plt.title('Feature Correlation heatmap')
plt.show()
```
